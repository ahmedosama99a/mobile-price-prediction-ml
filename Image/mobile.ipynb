# Data preprocessing

#### Data cleaning

```python
def cap_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    data[column] = data[column].apply(
        lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x)
    )

# تطبيقه على الأعمدة اللي فيها أوتلاير
cap_outliers(df, 'fc')
cap_outliers(df, 'px_height')
```

```python
outlier_cols = [ 'px_height', 'fc']

for col in outlier_cols:
    plt.figure(figsize=(6, 2))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot for {col}")
    plt.show()
```

```python
df['px_height'] = df['px_height'].replace(0, df['px_height'].median())
df['sc_w'] = df['sc_w'].replace(0, df['sc_w'].median())
```

#### Feature Selection

```python
x=df.drop('price_range',axis=1)
y=df['price_range']
```

#### Split Data

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Modeling

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

##### Logistic Regression

```python
log_model = LogisticRegression( max_iter=1000 , random_state=42)
log_model.fit(x_train_scaled,y_train)
log_pred = log_model.predict(X_test_scaled)
```

##### K-NN

```python
K_model = KNeighborsClassifier(n_neighbors=5)
K_model.fit(x_train_scaled,y_train)
K_pred = K_model.predict(X_test_scaled)
```

##### Decision Tree

```python
T_model = DecisionTreeClassifier() 
T_model.fit(x_train_scaled,y_train)
T_pred = T_model.predict(X_test_scaled)
```

# Evaluation 

```python
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix , ConfusionMatrixDisplay
```

```python
print('LogisticRegression Accuracy:', accuracy_score(y_test,log_pred))
print('KNN Accuracy:', accuracy_score(y_test,log_pred))
print('DecisionTree Accuracy:', accuracy_score(y_test,log_pred))
```

```python
print('LogisticRegression Accuracy:\n', classification_report(y_test,log_pred))
```

```python
features = ['ram', 'battery_power']
X_plot = df[features]
y = df['price_range']

# Scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_plot_scaled = scaler.fit_transform(X_plot)

# Split
from sklearn.model_selection import train_test_split
X_train_plot, X_test_plot, y_train_plot, y_test_plot = train_test_split(X_plot_scaled, y, test_size=0.2, random_state=42)

# Train model
from sklearn.neighbors import KNeighborsClassifier
knn_plot = KNeighborsClassifier(n_neighbors=11, metric='manhattan', weights='distance')
knn_plot.fit(X_train_plot, y_train_plot)

import numpy as np
import matplotlib.pyplot as plt

x_min, x_max = X_train_plot[:, 0].min() - 1, X_train_plot[:, 0].max() + 1
y_min, y_max = X_train_plot[:, 1].min() - 1, X_train_plot[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))
Z = knn_plot.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
scatter = plt.scatter(X_train_plot[:, 0], X_train_plot[:, 1], c=y_train_plot, cmap='viridis', edgecolor='k')
plt.xlabel('ram (scaled)')
plt.ylabel('battery_power (scaled)')
plt.title('Decision Boundary Using 2 Features (KNN)')
plt.legend(*scatter.legend_elements(), title="Price Range")
plt.show()
```
